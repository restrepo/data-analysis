```
import pandas as pd
import numpy as np
```

```
st=pd.read_csv("data.csv",names=["x1","x2","x3","x4","x5"])
```

```
st['sltn']=(st.x1+','+st.x2+','+st.x3+','+st.x4+','+st.x5).str.split(',')
st['sltnstr']=(st.x1+','+st.x2+','+st.x3+','+st.x4+','+st.x5)
```

## Solutions with 4 or 3 chiral fields

```
def dupli(x,a):
  isainx=False
  if a in x:
    isainx=True
  return isainx
```

```
st0=st[st.sltn.apply(lambda x: dupli(x,'0'))]
```

```
st0[st0.sltnstr.str.contains('0,0')]
```

```
st0[~st0.sltnstr.str.contains('0,0')]
```
